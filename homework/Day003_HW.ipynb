### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


```python
fh = open("./Data/xml/64_72hr_CH.xml",encoding = 'utf8')
f = fh.read()
fh.close()

print(f)
```

```python
import xmltodict

# 存取檔案

with open('./Data/xml/64_72hr_CH.xml',encoding = 'utf8') as fd:
    doc = dict(xmltodict.parse(fd.read()))

# 存取我們的資訊
print(doc['cwbopendata']['dataset']['locations']['locationsName'])
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？
arr=[]
# 用迴圈存取我們的資訊
chapters = doc['cwbopendata']['dataset']['locations']['location']
for chapter in chapters:
    arr.append(chapter['locationName'])
print (len(arr))
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

arr=[]
# 用迴圈存取我們的資訊
chapters = doc['cwbopendata']['dataset']['locations']['location']
for chapter in chapters:
    obj={}
    obj['地區']=chapter['locationName']
    obj['時間點']=chapter['weatherElement'][0]['time'][0]['dataTime']
    obj['溫度']=chapter['weatherElement'][0]['time'][0]['elementValue']['value']
    arr.append(obj)
print (arr)
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

arr=[]
# 用迴圈存取我們的資訊
chapters = doc['cwbopendata']['dataset']['locations']['location'][0]
local = chapters['locationName']
rows = chapters['weatherElement'][0]['time']
for row in rows:
    obj={}
    obj['地區']=local
    obj['時間點']=row['dataTime']
    obj['溫度']=row['elementValue']['value']
    arr.append(obj)
print (arr)
```
